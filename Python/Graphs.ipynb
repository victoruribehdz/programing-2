```python
import matplotlib.pyplot as plt
import pandas as pd

wktime1 = pd.read_csv('wktime.T001.his', delimiter = ' ', header = None)
plt.plot(wktime1[0], wktime1[1], color = 'b', marker = 'o', label = 'wktime1')
wktime4 = pd.read_csv('wktime.T004.his', delimiter = ' ', header = None)
plt.plot(wktime4[0], wktime4[1], color = 'c', marker = '*', label = 'wktime4')
plt.xlabel('w')
plt.ylabel('Probability p(w)')
plt.title("Probability Distribution")

plt.xscale("log")
plt.yscale("log")
plt.xlim(0.001, 100)
plt.ylim(0.000001, 100)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

scorers_x = ['Inmobile', 'Vardy', 'Messi', 'Mbappe', 'Aubameyang', 'Cristiano']
goals_1s = ['17', '12', '26', '18', '14', '20']
plt.plot(scorers_x, goals_1s, 'k-.', marker = 'o', label = 'season 1')

goals_2s = ['20', '11', '32', '24', '19', '27']
plt.plot(scorers_x, goals_2s, 'r--', marker = 'X', label = 'season 2')

goals_3s = ['25', '18', '24', '16', '17', '22']
plt.plot(scorers_x, goals_3s, 'b', marker = '*', label = 'season 3')

plt.legend()
plt.title("Maximum scorers in the last 3 seasons")
plt.xlabel("Scorers")
plt.ylabel("Goals")
plt.show()
```
